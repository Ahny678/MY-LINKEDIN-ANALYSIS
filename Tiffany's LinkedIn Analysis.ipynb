```python
import pandas as pd

df=pd.read_excel('/content/drive/MyDrive/TLP.xlsx')
```

```python
#editing row 3, comments column
df.at[2, 'Comments']=0
df
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set up the plot
plt.figure(figsize=(10,6))
sns.barplot(x='Impressions', y='Date', data=df, color='black')
plt.gca().set_facecolor('lightgreen')  # Inside the axes
plt.gcf().set_facecolor('green')

# Add labels and title
plt.title('Post Impressions vs Date')
plt.xlabel('Impressions')
plt.ylabel('Date')

plt.show()
```

```python
#get the unique values from the time feature for mapping
df['Time'].unique()
```

```python

import datetime

# Define the mapping dictionary
time_mapping = {
    datetime.time(8, 15): 'Morning',
    datetime.time(6, 59): 'Early Morning',
    datetime.time(9, 30): 'Morning',
    datetime.time(7, 54): 'Early Morning',
    datetime.time(8, 7): 'Morning',
    datetime.time(6, 39): 'Early Morning',
    datetime.time(23, 12): 'Night',
    datetime.time(17, 25): 'Evening',
    datetime.time(16, 26): 'Mid Afternoon',
    datetime.time(23, 20): 'Night',
    datetime.time(12, 19): 'Noon',
    datetime.time(17, 29): 'Evening',
    datetime.time(13, 38): 'Noon',
    datetime.time(11, 11): 'Morning',
    datetime.time(16, 30): 'Mid Afternoon'
}

# Step 2: Use replace() to map the times to time periods
df['time_period'] = df['Time'].replace(time_mapping)
df
```

```python
#  Plot the reactions against the time periods
plt.figure(figsize=(10, 6))
sns.barplot(x='time_period', y='Reactions', data=df, color='red')

# Add labels and title
plt.title('Reactions by Time of Day')
plt.xlabel('Time Period')
plt.ylabel('Reactions')

plt.show()
```

```python
#  Plot the reactions against the time periods
plt.figure(figsize=(10, 6))
sns.barplot(x='Type', y='Members Reached', data=df, color='purple')

plt.gca().set_facecolor('black')  # Inside the axes
plt.gcf().set_facecolor('lightgrey')

# Add labels and title
plt.title('Tyoe of Post vs Members Reached')
plt.xlabel('Type')
plt.ylabel('Members Reached')

plt.show()
```

```python
# Count plot for Top Jobs of viewers
plt.figure(figsize=(20,5))
sns.countplot(x='Top Job', data=df, color= 'black')

plt.gca().set_facecolor('pink')  # Inside the axes
plt.gcf().set_facecolor('lightgrey')

plt.title('Top Jobs of Viewers')
plt.xlabel('Top Jobs')
plt.ylabel('Count')

plt.show()
```
